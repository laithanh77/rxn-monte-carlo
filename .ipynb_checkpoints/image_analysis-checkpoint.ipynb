```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
from IPython.display import set_matplotlib_formats
import scipy.integrate as integrate
from matplotlib.pyplot import imshow
%matplotlib inline
set_matplotlib_formats('pdf', 'svg')

params = {'legend.fontsize': 'x-large',
          'figure.figsize': (12, 8),
         'axes.labelsize': 20,
         'axes.titlesize': 22,
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'mathtext.default':  'regular' 
         }
pylab.rcParams.update(params)


size = (10, 6)
img = Image.open("images/a_state_binary_clean.png")
img = img.convert("L")
img_arr = np.asarray(img)

imshow(np.uint8(255) - img_arr, cmap = "binary")
plt.axis("Off")
plt.title("Rough form of the potential curve in binary")


x = []
pairs = []
for i in range(img_arr.shape[0]):
    for j in range(img_arr.shape[1]):
        if img_arr[i][j] == 0 and j not in x:
            pairs.append((j, 227-i))
            x.append(j)
pairs = sorted(pairs, key = lambda x: (x[0]))

x = []
y = []
for i in pairs:
    x.append(i[0])
    y.append(i[1])

#extrapolate
if False:
    print("A")
    n = 100
    y += ([y[-1] for i in range(n-1)])
    x += ([i for i in range(x[-1]+1, x[-1]+n)])

x_arr = np.array(x)
y_arr = np.array(y)
x_scaled = ((x_arr/len(x_arr)) * 2.5)+2.55
y_scaled = (((y_arr/len(y_arr)) * 3.1)+1.25) *10e4



plt.figure(figsize=size)
plt.plot(x_scaled, y_scaled)
plt.title("Potential Energy Curve of I2 in excited A-State (Zewail 1995)")
plt.ylabel("Energy (cm^-1)")
plt.xlabel("Internuclear Separation (A)")
plt.xlim(2.5, 5)

h = 6.62607004e-32 #plancks constant
c = 299792458 #speed of light
m = 2.1072981e-25 #mass iodine kg


param = np.polyfit(x_scaled, y_scaled, 11)
fit = np.poly1d(param)
ypred = fit(x_scaled)

plt.figure(figsize=size)
plt.plot(x_scaled,y_scaled, label = "Energy")
plt.plot(x_scaled, ypred, label = "Fit")
plt.title("Potential Energy Curve Fitted")
plt.ylabel("Energy (cm^-1)")
plt.xlabel("Internuclear Separation (A)")
plt.xlim(2.5, 5)
plt.legend()

param_deriv = np.polyder(fit)
fit_deriv = np.poly1d(param_deriv)
force_cm = fit_deriv(x_scaled) * -1 #force is negative derivative of potential function, cm^-1

plt.figure(figsize=size)
force = force_cm*100*h*c #joules
plt.plot(x_scaled, force)
plt.plot(range(0,6), np.zeros_like(range(0,6)), "--", c = "black")
plt.title("Force vs internuclear separation")
plt.ylabel("Force (J)")
plt.xlabel("Internuclear separation (A)")
plt.xlim(2.5, 5)

acceleration = force/m
print("Velocity: {0} m/s (tranpezoid)".format(np.trapz(acceleration)))
velocity = integrate.quad(fit_deriv, a = 2.5, b = 5.0)[0]*100*h*c*-1*(1/m)
print("Velocity: {0} m/s (quadpack, more accurate probably)".format(velocity))
print(velocity * 1e-15 * 2.9116223400000000865 * 1e10, "angstroms per timestep")
```
